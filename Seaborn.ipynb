```
import seaborn as sns
import numpy as np
import pandas as pd
```

```
df=sns.load_dataset('tips')
df.head()
```

#####To find corelation between two feature

```
df.corr()
# we can se only int type features only in the corelation matrix
```

#HEAT MAP

```
#you can see the frequency of the coreletion
sns.heatmap(df.corr())
```

#JOINT MAP
######from this you can get two histograms and can decide which point is more focusable and you should focus

```
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex') 
```

```
 #change kind from hex to reg
 sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

#PAIR PLOT
######by this we will get all the combination of the different features

```
sns.pairplot(df)
```

```
sns.pairplot(df,hue='sex')
## we can differenciate male and female in this graphs.
```

#DIST PLOT
######it helps to check the distribution of the column

```
sns.distplot(df['tip'])
sns.distplot(df['tip'],kde=False)
```

#COUNT PLOT

```
sns.countplot('day',data=df)
```

#BAR PLOT

```
  sns.barplot(x='total_bill',y='sex',data=df)
```

#BOX-PLOT

```
sns.boxplot('day','total_bill',data=df)
```

```
sns.boxplot('sex','total_bill',data=df)
```

#VIOLIN PLOT

```
sns.violinplot(x='day',y='total_bill',data=df)
```

```
sns.violinplot(x='total_bill',y='sex',data=df)
```
