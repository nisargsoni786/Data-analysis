```
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```
x=np.arange(0,10)
y=np.arange(11,21)
a=np.arange(40,50)
b=np.arange(50,60)
```

```
plt.scatter(x,y,c='g')
plt.xlabel('x axis')
plt.ylabel('y-axis')
plt.title('PLOTTING')
```

```
y=x*x
#'r-': red line
#'r*':red star          'ro":red dots like scatter 
#'r-*': line and stars at point
plt.plot(x,y,'r*--',markersize=15)
```

```
plt.subplot(2,2,1)
plt.plot(x,y,'o')
plt.subplot(2,2,4)
plt.plot(x,x*x*x,'r--*')
plt.subplot(2,2,2)
plt.plot(x,x*x*2+9,'g*-',markersize=12)
```

```
x=np.arange(0,4*(np.pi),0.3)
plt.plot(x,np.sin(x),'r-*')
```

#BOX-PLOT

```
data=[np.random.normal(0,i,100) for i in range(1,4)]
data
```

```
#patch artist is for color inside the box.
#vert is for vertical or horizontal...
plt.boxplot(data,vert=True,patch_artist=True);
```

```
a=np.array([[-2,1,2,9],[4,5,6],[7,8,9]])
plt.boxplot(a,vert=False,patch_artist=True);
```

#pie-chart

```
label=['python','c++','ruby','java']
sizes=[215,130,245,210]
colors=['gold','lightgreen','lightcoral','skyblue']
explode=(0.1,0.0,0.0,0)

# Now plot

plt.pie(sizes,explode=explode,labels=label,colors=colors,autopct='%1.1f%%',shadow=True)
plt.axis('equal')


#python is bit outside bcz we have explode it....
```
